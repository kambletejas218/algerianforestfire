# DATA CLEANING

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

## The Dataset is converted  into two sets  based on region from 122th index and we can make a new column based on the region "

## 1 : Bejaia Region Dataset 
## 2 : Sidi-Bel Abbes Region Dataset
### Add new column with region

```python
df.loc[:122,'Region'] = 0
df.loc[122: ,'Region'] = 1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## 122 index which has only column name their is no data
df =df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns name 
df.columns = df.columns.str.strip()
df.columns
```

## Change the required column as int

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

## Change the required column as float

```python
df[[ 'Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[[ 'Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

## lets save the cleaned dataset

```python
df.to_csv("Algerian_cleaned.csv",index=False) 
```
